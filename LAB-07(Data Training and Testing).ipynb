```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pearsonr

# Given data
study_hours = np.array([2, 4, 6, 8, 1, 5, 7, 3, 9, 10])
exam_scores = np.array([50, 60, 70, 85, 40, 65, 78, 55, 90, 95])
sleep_hours = np.array([8, 7, 6, 5, 9, 6.5, 5.5, 7.5, 4, 3])

# Function to create scatter plots
def create_scatter_plot(x, y, x_label, y_label, title):
    plt.scatter(x, y, color='blue', edgecolors='black', alpha=0.7)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.title(title)
    plt.grid(True)
    plt.show()

# Scatter Plots
create_scatter_plot(study_hours, exam_scores, "Study Hours", "Exam Scores", "Study Hours vs Exam Scores")
create_scatter_plot(sleep_hours, exam_scores, "Sleep Hours", "Exam Scores", "Sleep Hours vs Exam Scores")
create_scatter_plot(study_hours, sleep_hours, "Study Hours", "Sleep Hours", "Study Hours vs Sleep Hours")

# Calculate Pearson correlation coefficients
corr_study_exam, _ = pearsonr(study_hours, exam_scores)
corr_sleep_exam, _ = pearsonr(sleep_hours, exam_scores)
corr_study_sleep, _ = pearsonr(study_hours, sleep_hours)

# Print correlation coefficients
print(f"Pearson correlation (Study Hours & Exam Scores): {corr_study_exam:.3f}")
print(f"Pearson correlation (Sleep Hours & Exam Scores): {corr_sleep_exam:.3f}")
print(f"Pearson correlation (Study Hours & Sleep Hours): {corr_study_sleep:.3f}")
```
