```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score
data={
    'TV':[20,25,22,27],
    'Radio':[10,12,11,14],
    'Social Media':[15,17,16,18],
    'Sales':[50,70,65,80],
}
df=pd.DataFrame(data)
#split the data
x=df[['TV','Radio','Social Media']]
y=df['Sales']
# Training the model
model=LinearRegression()
model.fit(x,y)

#extracting coefficients
intercept=model.intercept_
coefficients=model.coef_
print(f"Intercept:{intercept}")
print(f"coefficients:{coefficients}")


```

```python
#model evaluation
#model evaluation
y_pred = model.predict(x)
r2=r2_score(y,y_pred)
mse=mean_squared_error(y,y_pred)
rmse=np.sqrt(mse)
print(f"R-squared:{r2}")
print(f"Mean squared error:{mse}")
print(f"Root mean squared error:{rmse}")
```

```python
# Visualization
plt. figure(figsize=(8, 6))
# Adding the predicted sales to the dataframe
df['Predicted Sales'] = y_pred  # Assign the predicted values to a new column 'Predicted Sales'
sns.scatterplot(x=df['Sales'], y=df['Predicted Sales'])
plt.plot([df['Sales'].min(), df['Sales'].max()], [df['Sales'].min(), df['Sales'].max()], 'p--')
plt.xlabel( 'Actual Sales')
plt.xlabel( 'Predicted Sales')
plt.title('Actual vs. Predicted Sales')
```
