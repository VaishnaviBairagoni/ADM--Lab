```python
import numpy as np

x=np.array([8,26,48,36,54,120,162,28])
y=np.array([55,69,88,22,434,143,431,222])
print(x)
print(y)

#split the data into trainong (80)% and testing (20)%
#first values for training
x_train=x[:8]
y_train=y[:8]

#last 2 values for testing
x_test=x[:8]
y_test=y[:8]

#print the dataset
print("Training Data (x):",x_train)
print("Training Data (y):",y_train)
print("Testing  Data (x):",x_test)
print("Testing Data (y):",y_test)

```

```python
#compute means
x_mean=np.mean(x_train)
y_mean=np.mean(y_train)

#compute slope (bet 1)
numerator=np.sum((x_train-x_mean)*(y_train-y_mean))
denominator=np.sum((x_train-x_mean)**2)
beta_1=numerator/denominator
print("Intercept(beta_1):",beta_1)

#compute intercept (beta_0)
beta_0=y_mean-beta_1*x_mean
print("Slope (beta_0):",beta_0)
```

```python
#predict y values using the regression equation testing data
y_pred=beta_0+beta_1*x_test
print("Predicted y vales:",y_pred)
```

```python
#compute thr square root errors(MSE) for evaluation
mse=np.mean((y_test-y_pred)**2)
#compute R-squared value
ss_total=np.sum((y_test-np.mean(y_test))**2)
ss_residual=np.sum((y_test-y_pred)**2)
r_squared=1-(ss_residual/ss_total)
print("Mean Squared error (MSE):",mse)
print("R-squared:",r_squared)
```

```python
#visualation
import matplotlib.pyplot as plt
plt.scatter(x_train,y_train,color='blue',label='Training Data')
plt.scatter(x_test,y_test,color='red',label='Testing Data')
plt.plot(x_test,y_pred,color='green',linestyle='dashed',label='Regression Line')
plt.xlabel("Hours Worked")
plt.ylabel("Business Growth")
plt.legend()
plt.title("Simple Linear Regression ")
plt.show()
```
